```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import cv2
import numpy as np

%matplotlib inline
```

```python
image = mpimg.imread('curved_lane.jpg')
plt.imshow(image)
```

## Converting image to grayScale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
filter_y = np.array([[-1, -2, -1], 
                    [ 0, 0, 0], 
                    [ 1, 2, 1]])

filter_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

filtered_imageY = cv2.filter2D(gray_image, -1, filter_y)
plt.imshow(filtered_imageY, cmap='gray')
```

```python
filtered_imageX = cv2.filter2D(gray_image, -1, filter_x)
plt.imshow(filtered_imageX, cmap='gray')
```

### I will try some random Filters

```python
filter_1 = np.array([[1, 0, 1], 
                    [ 0, 0, 0], 
                    [ 1, 0, 1]])
img1 = cv2.filter2D(gray_image, -1, filter_1)
plt.imshow(img1, cmap='gray')
```

```python
filter_1 = np.array([[0, 0, 0], 
                    [ 0, 1, 0], 
                    [ 0, 0, 0]])
img1 = cv2.filter2D(gray_image, -1, filter_1)
plt.imshow(img1, cmap='gray')
```

```python
filter_1 = np.array([[1/9, 1/9, 1/9], 
                    [ 1/9, 1/9, 1/9], 
                    [ 1/9, 1/9, 1/9]])
img1 = cv2.filter2D(gray_image, -1, filter_1)
plt.imshow(img1, cmap='gray')
```

```python
filter_1 = np.array([[1/9, 1/9, 1/9], 
                    [ 1/9, 1/9, 1/9], 
                    [ 1/9, 1/9, 1/9]])
img1 = cv2.filter2D(gray_image, -1, filter_1)
plt.imshow(img1, cmap='gray')
```

```python
kernel = np.ones((20, 20)) / (20*20)
img1 = cv2.filter2D(gray_image, -1, kernel)
plt.imshow(img1, cmap='gray')
```

```python
filter_1 = np.array([[1/9, 1/9, 1/9], 
                    [ 1/9, 20, 1/9], 
                    [ 1/9, 1/9, 1/9]])
img1 = cv2.filter2D(gray_image, -1, filter_1)
plt.imshow(img1, cmap='gray')
```

```python
kernel = np.zeros((100, 100))
kernel[0][0] = 1
img1 = cv2.filter2D(gray_image, -1, kernel)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(gray_image, cmap='gray')
axs[0].set_title("Original Image")
axs[1].imshow(img1, cmap='gray')
axs[1].set_title("Filtered Image")
```
